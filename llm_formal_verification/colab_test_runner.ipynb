```python
# 1. Install System Dependencies (Why3 and Alt-Ergo via OPAM)
# why3 and alt-ergo are NOT in Colab's apt repos, so we install via OPAM
!apt-get update > /dev/null 2>&1
!apt-get install -y opam m4 build-essential libgmp-dev > /dev/null 2>&1
print("✅ Base system packages installed.")

# Initialize OPAM (non-interactive)
!opam init --disable-sandboxing --auto-setup -y > /dev/null 2>&1
print("✅ OPAM initialized.")

# Install why3 and alt-ergo via OPAM
!eval $(opam env) && opam install -y why3 alt-ergo 2>&1 | tail -5
print("✅ Why3 and Alt-Ergo installed via OPAM.")

# Add OPAM bin to PATH for this session
import os
opam_bin = os.path.expanduser("~/.opam/default/bin")
os.environ["PATH"] = opam_bin + ":" + os.environ["PATH"]

# Configure why3 to detect alt-ergo
!why3 config detect 2>&1 | tail -3

# Verify installation
!which why3 && why3 --version
!which alt-ergo && alt-ergo --version
print("✅ Why3 and Alt-Ergo are ready.")
```

```python
# 2. Clone or Update the Repository
import os

REPO_URL = "https://github.com/Vtheonly/RAG_LLM/"
REPO_NAME = "RAG_LLM"
PROJECT_DIR = "llm_formal_verification"

if os.path.exists("/content"):
    %cd /content

if os.path.exists(REPO_NAME):
    print(f"Repository '{REPO_NAME}' already exists. Pulling latest updates from GitHub...")
    %cd {REPO_NAME}
    !git pull
    %cd {PROJECT_DIR}
else:
    print(f"Cloning {REPO_URL}...")
    !git clone {REPO_URL} {REPO_NAME}
    %cd {REPO_NAME}/{PROJECT_DIR}

print("✅ Codebase locally synced.")
```

```python
# 3. Install Python Dependencies
# Ensure OPAM bin stays in PATH for subprocess calls from Python
import os
opam_bin = os.path.expanduser("~/.opam/default/bin")
os.environ["PATH"] = opam_bin + ":" + os.environ["PATH"]

!pip install -r requirements.txt > /dev/null 2>&1
print("✅ Python prerequisites installed.")
```
