This notebook is the cloud-companion to your formal verification project. It automates the installation of Why3 and Alt-Ergo using the **OPAM** package manager to ensure maximum stability.

**Recommended Runtime**: GPU (T4 or L4) for faster inference.

```python
# 1. Install System Dependencies & OPAM
!apt-get update > /dev/null 2>&1
!apt-get install -y opam m4 build-essential libgmp-dev > /dev/null 2>&1
print("✅ Base system packages installed.")

# Initialize OPAM (non-interactive)
!opam init --disable-sandboxing --auto-setup -y > /dev/null 2>&1
print("✅ OPAM initialized.")

# Install why3 and alt-ergo via OPAM (This takes ~5-10 minutes)
print("⏳ Installing Why3 and Alt-Ergo... please wait.")
!eval $(opam env) && opam install -y why3 alt-ergo > /dev/null 2>&1
print("✅ Why3 and Alt-Ergo installed via OPAM.")

# Add OPAM bin to PATH for this session
import os
opam_bin = os.path.expanduser("~/.opam/default/bin")
os.environ["PATH"] = opam_bin + ":" + os.environ["PATH"]

# Configure why3 to detect alt-ergo
!why3 config detect
print("✅ Why3 configured and solvers detected.")
```
